```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['font.weight'] = 'medium'

base_path = '/home/dell/Desktop/BA/' 

try:
    df_org = pd.read_csv(os.path.join(base_path, 'review.csv'))
    df_filt = pd.read_csv(os.path.join(base_path, 'review_filtered.csv'))
    
    df_org['Source'] = 'Raw Dataset'
    df_filt['Source'] = 'Filtered Dataset'
    
    df_all = pd.concat([df_org, df_filt], ignore_index=True)
    
    meta_path = os.path.join(base_path, 'metadata.csv')
    if os.path.exists(meta_path):
        df_meta = pd.read_csv(meta_path)
        cols_to_keep = [col for col in ['parent_asin', 'main_category', 'title'] if col in df_meta.columns]
        df_meta = df_meta[cols_to_keep]
    else:
        df_meta = None
        print("[WARNING] Metadata file not found. Skipping category analysis.")

except FileNotFoundError:
    print(f"[ERROR] Critical files missing at: {base_path}")
    exit()

def calculate_metrics(df, name):
    n_users = df['user_id'].nunique()
    n_items = df['parent_asin'].nunique()
    n_interactions = len(df)
    
    avg_inter_per_user = n_interactions / n_users if n_users > 0 else 0
    matrix_size = n_users * n_items
    sparsity = 1 - (n_interactions / matrix_size) if matrix_size > 0 else 0
    
    return {
        'Dataset Version': name,
        'Unique Users': n_users,
        'Unique Items': n_items,
        'Total Interactions': n_interactions,
        'Avg Interactions/User': round(avg_inter_per_user, 2),
        'Sparsity Level (%)': round(sparsity * 100, 4)
    }

metrics_org = calculate_metrics(df_org, 'Raw Dataset')
metrics_filt = calculate_metrics(df_filt, 'Filtered Dataset')

df_metrics = pd.DataFrame([metrics_org, metrics_filt])

print("     REPORT: DATASET OVERVIEW & SPARSITY METRICS       ")
print(f"Retention Rate: {len(df_filt)/len(df_org)*100:.2f}% of original data kept.")
print("-" * 55)
print(df_metrics.to_string(index=False))
print("-" * 55)

def plot_user_analysis():
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    
    user_counts_org = df_org['user_id'].value_counts()
    user_counts_filt = df_filt['user_id'].value_counts()
    
    sns.kdeplot(user_counts_org, fill=True, label='Raw Dataset', color='#e74c3c', alpha=0.3, clip=(0, 100), ax=axes[0])
    sns.kdeplot(user_counts_filt, fill=True, label='Filtered Dataset', color='#3498db', alpha=0.3, clip=(0, 100), ax=axes[0])
    axes[0].set_title('Figure 1: User Interaction Density Estimation (KDE)')
    axes[0].set_xlabel('Number of Interactions')
    axes[0].set_ylabel('Density')
    axes[0].set_xlim(0, 50)
    axes[0].legend()
    
    sns.histplot(user_counts_org, bins=50, color='#e74c3c', alpha=0.5, label='Raw Dataset', ax=axes[1], log_scale=(False, True))
    sns.histplot(user_counts_filt, bins=50, color='#3498db', alpha=0.6, label='Filtered Dataset', ax=axes[1], log_scale=(False, True))
    
    axes[1].axvline(x=3, color='black', linestyle='--', linewidth=1.5, label='Filter Threshold (k=3)')
    axes[1].set_title('Figure 2: Interaction Distribution (Log Scale) & Threshold')
    axes[1].set_xlabel('Number of Reviews per User')
    axes[1].set_ylabel('Count of Users (Log Scale)')
    axes[1].set_xlim(0, 50)
    axes[1].legend()

    plt.tight_layout()
    plt.show()

def plot_item_analysis():
    plt.figure(figsize=(12, 6))
    
    def get_pareto_data(df):
        item_counts = df['parent_asin'].value_counts().reset_index()
        item_counts.columns = ['parent_asin', 'count']
        item_counts = item_counts.sort_values(by='count', ascending=False)
        item_counts['cumulative_perc'] = item_counts['count'].cumsum() / item_counts['count'].sum()
        item_counts['item_rank_perc'] = (np.arange(len(item_counts)) + 1) / len(item_counts)
        return item_counts
    
    pareto_org = get_pareto_data(df_org)
    pareto_filt = get_pareto_data(df_filt)
    
    plt.plot(pareto_org['item_rank_perc'], pareto_org['cumulative_perc'], label='Raw Dataset', color='#e74c3c')
    plt.plot(pareto_filt['item_rank_perc'], pareto_filt['cumulative_perc'], label='Filtered Dataset', color='#3498db', linestyle='--')
    plt.axhline(y=0.8, color='gray', linestyle=':', label='80% Cumulative Interactions')
    
    plt.title('Figure 3: Pareto Principle Analysis (Long-tail Distribution)')
    plt.xlabel('Percentage of Items (Ranked by Popularity)')
    plt.ylabel('Percentage of Cumulative Interactions')
    plt.legend()
    plt.grid(True, which="both", ls="-", alpha=0.5)
    plt.show()
    
    if df_meta is not None and 'main_category' in df_meta.columns:
        plt.figure(figsize=(12, 6))
        merged_df = pd.merge(df_filt, df_meta, on='parent_asin', how='inner')
        top_cats = merged_df['main_category'].value_counts().head(10)
        
        sns.barplot(x=top_cats.values, y=top_cats.index, palette='viridis')
        plt.title('Figure 4: Top 10 Dominant Product Categories (Filtered Dataset)')
        plt.xlabel('Total Review Count')
        plt.ylabel('Category')
        plt.show()

def plot_rating_distribution():
    plt.figure(figsize=(10, 6))
    
    rating_counts = df_all.groupby(['Source', 'rating']).size().reset_index(name='count')
    total_counts = rating_counts.groupby('Source')['count'].transform('sum')
    rating_counts['percentage'] = rating_counts['count'] / total_counts * 100
    
    sns.barplot(data=rating_counts, x='rating', y='percentage', hue='Source', palette={'Raw Dataset': '#e74c3c', 'Filtered Dataset': '#3498db'})
    
    plt.title('Figure 5: Rating Distribution & Bias Check')
    plt.ylabel('Percentage (%)')
    plt.xlabel('Rating Score')
    plt.legend(title='Dataset Source')
    plt.show()

def process_date_smart(df):
    if 'timestamp' not in df.columns: return None
    try:
        if pd.api.types.is_numeric_dtype(df['timestamp']):
            unit = 'ms' if df['timestamp'].mean() > 1e11 else 's'
            return pd.to_datetime(df['timestamp'], unit=unit)
        return pd.to_datetime(df['timestamp'], errors='coerce')
    except:
        return pd.to_datetime(df['timestamp'], errors='coerce')

def plot_time_trend():
    plt.figure(figsize=(14, 6))
    
    df_org['date'] = process_date_smart(df_org)
    df_filt['date'] = process_date_smart(df_filt)
    
    if df_org['date'].isna().all():
        return 

    trend_org = df_org.set_index('date').resample('M').size()
    trend_filt = df_filt.set_index('date').resample('M').size()
    
    plt.plot(trend_org.index, trend_org.values, label='Raw Dataset', color='#e74c3c', alpha=0.5)
    plt.plot(trend_filt.index, trend_filt.values, label='Filtered Dataset', color='#3498db', linewidth=2)
    
    plt.title('Figure 6: Temporal Interaction Trends (Time-Drift Analysis)')
    plt.xlabel('Timeline')
    plt.ylabel('Monthly Interactions')
    plt.legend()
    plt.show()

plot_user_analysis()
plot_item_analysis()
plot_rating_distribution()
plot_time_trend()
```

```python
plot_user_analysis()
plot_item_analysis()
plot_rating_distribution()
plot_time_trend()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os


def load_data(file_path, model_name):
    if not os.path.exists(file_path): return None
    try:
        df = pd.read_csv(file_path)
        df['Model'] = model_name
        if 'actual_items' in df.columns: df.rename(columns={'actual_items': 'groundtruth_items'}, inplace=True)
        
        def count_items(x):
            if pd.isnull(x): return 0
            s = str(x).replace('|', ',')
            return len([i for i in s.split(',') if i.strip() != ''])

        df['cnt'] = df['groundtruth_items'].apply(count_items)
        df = df[df['cnt'] <= 15] 
        
        bins = [0, 5, 16]
        labels = ['Small (<5)', 'Medium (5-15)']
        df['Group'] = pd.cut(df['cnt'], bins=bins, labels=labels, include_lowest=True)
        return df
    except: return None

files = [
    ('/home/dell/Desktop/BA/results/knn.csv', 'ItemKNN'),
    ('/home/dell/Desktop/BA/results/blair.csv',   'BLAIR'),
    ('/home/dell/Desktop/BA/results/lightgcl.csv', 'LightCGL')
]
dfs = [load_data(p, n) for p, n in files if load_data(p, n) is not None]
df_all = pd.concat(dfs, ignore_index=True) if dfs else None

custom_palette = ['#95a5a6', '#34495e', '#e74c3c'] 
sns.set_theme(style="white", font_scale=1.2)

def plot_bar_chart():
    if df_all is None: return
    
    metrics = ['recall@10', 'ndcg@10', 'recall@50', 'ndcg@50']
    df_melt = df_all.melt(id_vars=['Model'], value_vars=metrics, var_name='Metric', value_name='Score')

    plt.figure(figsize=(10, 6))
    
    ax = sns.barplot(
        data=df_melt, x='Metric', y='Score', hue='Model', 
        palette=custom_palette, edgecolor='white', linewidth=1, errorbar=None
    )
    
    for container in ax.containers:
        ax.bar_label(container, fmt='%.3f', padding=3, fontsize=10, fontweight='bold')

    plt.title('Overall Performance Comparison', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('')
    plt.ylabel('Score')
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    plt.legend(loc='upper left', frameon=False)
    
    sns.despine(left=True)
    plt.tight_layout()
    plt.show()

def plot_line_chart():
    if df_all is None: return

    plt.figure(figsize=(8, 6))
    
    sns.lineplot(
        data=df_all, x='Group', y='ndcg@10', hue='Model', style='Model',
        palette=custom_palette, 
        markers=True, dashes=False, 
        linewidth=4, markersize=12,
        errorbar=None
    )
    
    plt.title('Robustness Analysis: NDCG@10 by Basket Size', fontsize=16, fontweight='bold', pad=20)
    plt.ylabel('NDCG@10 Score')
    plt.xlabel('User Group')
    
    plt.ylim(bottom=0)
    plt.grid(True, linestyle='--', alpha=0.4)
    
    plt.legend(title='Model Architecture', loc='upper right', frameon=True)
    sns.despine()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("\n📊 Đang vẽ Bar Chart...")
    plot_bar_chart()
    
    print("\n📈 Đang vẽ Line Chart...")
    plot_line_chart()
```
